```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Lable')
plt.title('Title')
```

```python
plt.show()
```

```python
fig =plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,color='purple',lw=3,ls='-')
axes.set_xlabel('X lable')
axes.set_ylabel('Y lable')
axes.set_title('my title')
plt.show()
```
